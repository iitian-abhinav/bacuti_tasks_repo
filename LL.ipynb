# Singly Lined List

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
class SLL:
    def __init__(self):
        self.head=None

    def traversal(self):
        if self.head is None:
            print("SLL is empty")
        else:
            a=self.head
            while a is not None:
                print(a.data, end=",")
                a=a.next

```

```python
n1=Node(5)
sll=SLL()
sll.head=n1
n2=Node(10)
n1.next=n2
n3=Node(15)
n2.next=n3
n4=Node(20)
n3.next=n4
n5=Node(25)
n4.next=n5
n6=Node(30)
n5.next=n6
```

```python
sll.traversal()
```

### Insertion

```python
# Insertion at beginning
def insertAtBeg(self,data):
    nb=Node(data)
    nb.next=self.head
    self.head=nb
```

```python
insertAtBeg(sll,2)
sll.traversal()
```

```python
def insertAtEnd(self,data):
    a=self.head
    while a.next is not None:
        a=a.next
    ne=Node(data)
    a.next=ne
```

```python
insertAtEnd(sll,32)
sll.traversal()
```

```python
def insertAtPosition(self,data,position):
    nib=Node(data)
    a=self.head
    for i in range(1,position-1):
        a=a.next
    nib.next=a.next
    a.next=nib
```

```python
insertAtPosition(sll,4,6)
sll.traversal()
```

### Deletion

```python
def deleteAtBeg(self):
    if self.head is None:
        return
    a = self.head
    self.head = a.next
    a.next = None
```

```python
deleteAtBeg(sll)
sll.traversal()
```

```python
def deleteAtEnd(self):
    prev=self.head
    a=self.head.next
    while a.next is not None:
        a=a.next
        prev=prev.next
    prev.next=None
```

```python
deleteAtEnd(sll)
sll.traversal()
```
